# Fitting a line with scipy

```python
### Define function performing regression and creating plot

def linear(x_data,y_data,xlabel,ylabel):
    
    import numpy as np
    from scipy import stats
    import matplotlib.pyplot as plt

    m1,b1,r1,p1,se1 = stats.linregress(x_data,y_data)

    ### Generate line of best fit between limits of x_dat
    x_fit = np.linspace(np.min(x_data),np.max(x_data),101)
    y_fit = m1 * x_fit + b1

    ### Generate an example plot for fit
    fig = plt.figure(figsize=(2.5,2.5), dpi=150)
    ax1 = plt.subplot(111)

    ax1lab="$"+('y=%2.2fx+%2.2f, r^2=%1.2f'%(m1,b1,r1**2))+"$"
    ax1.scatter(x_data,y_data,marker='o',edgecolors='black')
    ax1.plot(x_fit,y_fit,'--',label=ax1lab)
    ax1.legend(loc='upper left',fontsize=7,frameon=False)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)

    plt.show()
   
```

```python
### Example usage

import numpy as np

k = np.array([0.013164,0.017159564,0.026766312,0.021206601,0.05818212])
T = np.array([30,35,40,45,50])+273

linear(1/T,np.log(k),'1/T (1/K)','ln(rate)')
```
