# Quick plot

```python
###### Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import signal
from scipy.optimize import curve_fit
from scipy.stats.distributions import t as tdist
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from scipy.interpolate import CubicSpline
```

```python
###### Import data

data_df = pd.read_excel('line_data.xlsx')
xdat=data_df['x'].to_numpy()
ydat=data_df['y'].to_numpy()

###### Plot peak fits
fig = make_subplots(1,1)

# Data
fig.add_trace(go.Scatter(x=xdat,y=ydat,name='Data'))
    
### Changes Gridlines, Add Axis Titles, Add Border, White background
fig.update_xaxes(title_text='<b>x<b>',gridcolor='black',row=1,col=1)
fig.update_xaxes(showline=True,linewidth=3,linecolor='black',mirror=True,row=1,col=1)
fig.update_yaxes(title_text='<b>y<b>',gridcolor='black',row=1,col=1)
fig.update_yaxes(showline=True,linewidth=3,linecolor='black',mirror=True,row=1,col=1)
fig.update_layout(
    plot_bgcolor='white',
    width=700,
    height=500,
    autosize=False,
    showlegend=False,
    font_size = 24,
    font_color = 'black',
    font_family = 'Arial',
)
```
