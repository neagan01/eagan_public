## The directory

Since you will be calling files, it is important here to note the importance of directory location. If you are not in the directory (or folder) where your file is, you either need to move to that directory or specify the file path when you run it. You can change directories wither in the graphical interface or with the "cd" command.

## Running python files

In Spyder (or similar software), the best way to run .py files is with the "runfile" command. Consider file "test.py" located in "C:/Users/Username/Documents/Python_files". If you are already in this directory, you can run the file via:

    runfile('test.py')

If you are not, you should instead run:

    runfile('C:/Users/Username/Documents/Python_files/test.py', wdir='C:/Users/Username/Documents/Python_files')

In Spyder there is also a run button (right-facing triangle) you can press instead if you prefer.

Pressing "shift-enter" will also run the current *cell* you are in, like in a Jupyter notebook. Here "cells" are separated with:

    # %%

If you are instead in the terminal, you instead run the file by typing:

    python test.py

Note that directory rules still apply and that you must first load your environment in this case. To do this, you will have to type:

    C:/Users/USERNAME/anaconda3/Scripts/activate

If you wish to activate another environment - perhaps a python2 environment called 'py2' - you could then type:

    conda activate py2

If you wish to return to your base environment, you could return with:

    conda activate base

## Plotting

Plots require a graphical interface because they are, well, graphics. This makes rendering plots straight from the command prompt tricky unless you download something like "kaleido" to use alongside "plotly," which generates the images in a browser window by default.

 In Jupyter, you sometimes need to type "fig.show()" or "plt.show()" to make your figure appear.
 
 In Spyder, you do not need to add this command - doing so many actually trigger an error.


## General piece of advice

If you wish to create your python files in an editor like Spyder because you prefer that mode of coding, you may do so. It is still better to then copy this file's contents and past it into a single Jupyter cell in order to group inputs and outputs together.

## Example

Copy the cell below, save it as "test.py" and then run it in Spyder. Depending on how you run it, either all cells or one cell only will be evaluated.

Note that you can also just run it below.

```python
# %% Cell 1

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,2.5),dpi=300)
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
ax2b = ax2.twinx()

x4 = [1,2,3,4]
y4 = [3,4,5,6]
y4b = [100,70,30,10]

ax1.plot(x4,y4,'-.c',label='Data')
ax1.set_xlabel('X4 label (units)')
ax1.set_ylabel('Y4 label (units)')
ax1.legend()

ax2.plot(x4,y4,'--c')
ax2.scatter(x4,y4,marker='o',edgecolors= 'black',label='Data',color='c')
ax2.yaxis.label.set_color('c')
ax2.set_xlabel('X4 label (units)')
ax2.set_ylabel('Y4 label (units)')
ax2.tick_params(axis='y', colors='c')
ax2.set_xlim([-0.5,4.5])

ax2b.plot(x4,y4b,':r')
ax2b.scatter(x4,y4b,marker='p',edgecolors='purple',label='Other Data',color='r')
ax2b.set_ylabel(r'Y4b ($\alpha^2$)')
ax2b.yaxis.label.set_color('r')
ax2b.tick_params(axis='y', colors='r')

fig.tight_layout()

print('Cell 1 run')

# %% Cell 2
print('Cell 2 run')
```
