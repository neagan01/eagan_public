## Defining functions

Oftentimes we need to create our own functions to do things. They can take zero, one, or more inputs. Here we define a function which squares an input and adds 12:

```python
def fun1(input):
    output = input**2 + 12
    return output
```

Now in a separate cell (or in the same cell) we can run this:

```python
fun1(16)
```

We can store this output as a new variable too:

```python
newvar = fun1(16)
print(newvar)
```

Note that if you had not evaluated the cell where the function is defined, this would return an error.

Now we can define a new function which takes in two arrays and a scalar. It will add thw two arrays together and then raise the final array to the scalar value.

Note that we use numpy arrays here. We will define it in this cell in case it has not already been in here.

```python
def fun2(inp1,inp2,inp3):
    output = (inp1+inp2)**inp3
    return output

import numpy as np
fun2(np.array([1,2,3,4]),np.array([3,4,5,6]),7)
```

Note that if we include the wrong data types, an error will be thrown. Let's see what happens when we use lists instead of numpy arrays:

```python
fun2([1,2,3,4],[3,4,5,6],7)
```

If we want, we can make the function convert the input to a numpy array:

```python
def fun3(inp1,inp2,inp3):
    import numpy as np
    inp1 = np.array(inp1)
    inp2 = np.array(inp2)
    output = (inp1+inp2)**inp3
    return output

fun3([1,2,3,4],[3,4,5,6],7)
```

This works regardless of input type, as long as the shapes match. Here's what happens if the arrays don't match sizes:

```python
fun3([1,2,3],[3,4,5,6],7)
```

**This is a very common error you will run into!**

## Lambda functions

Small functions can be written more concisely with lambda functions. Here the syntax is:

    lambda arguments : expression

For example, we can create function z1 which adds 2 to argument a1:

```python
z1 = lambda a1 : a1 + 2
print(z1(2))
```

We can also make this a function of multiple variables. Here we add the first two scalars and raise them to the third scalar:

```python
z2 = lambda a1, a2, a3 : (a1+a2)**a3
print(z2(5,4,3))
```

These are particularly useful when incorporating them into larger functions without having to define one elsewhere and track it down later.

### Scipy
Various algorithms useful for problem solving are found in scipy. These include non-linear systems of equations solvers, non-linear regression, and ODE solvers.

#### fsolve: systems of equations solver

"opt.fsolve" is the method used here. To run this you need (1) a system of equations and (2) guess values. The system of equations will be input as a function which takes the unknown variable as inputs and outputs one value for each equation which should be equal to zero when solved. An example is shown below.

This example aims to solve the following set of equations:

$$ y_1^2 + 2 \cdot y_1 = 10 $$
$$ 2 \cdot y_1^2 + y_2 = 12 $$
$$ 2 + y_3 = 8 $$

We will rearrange the equations above so that one side is simply 0, guess values of the parameters, run the solver, print the solutions, and then check that the solver correctly set all equations equal to zero.

```python
###### Import numpy and solver
import numpy as np
import scipy.optimize as opt

###### Define equations
def eqns(U):
    
    # Redefine inputs
    y1 = U[0]
    y2 = U[1]
    y3 = U[2]
    
    # Define equations
    setzero = np.zeros(len(U))
    setzero[0] = y1**2 + 2 * y1 - 10
    setzero[1] = 2 * y1**2 + y2 - 12
    setzero[2] = 2 + y3 - 8
    
    # Reorganize list to fit required dimensions 
    setzero = np.array(setzero).tolist()
    
    return setzero

###### Guess values
y1_guess = 1
y2_guess = 1
y3_guess = 1
guess = [y1_guess,y2_guess,y3_guess]

####### Run solver
soln = opt.fsolve(eqns,guess)

y1_solved = soln[0]
y2_solved = soln[1]
y3_solved = soln[2]

print('y1 = ','{0:,.4f}'.format(y1_solved))
print('y2 = ','{0:,.4f}'.format(y2_solved))
print('y3 = ','{0:,.4f}\n'.format(y3_solved))  

### Check solution
soln_check = eqns(soln)
print('Eqn 1 should be 0 and it is:','{0:,.5f}'.format(soln_check[0]))
print('Eqn 2 should be 0 and it is:','{0:,.5f}'.format(soln_check[1]))
print('Eqn 3 should be 0 and it is:','{0:,.5f}'.format(soln_check[2]))
```

### curve_fit: non-linear regression

Non-linear regression can be performed using "curve_fit." Here you define a model which has a set of "input" parameters you are solving for. Then you will guess these parameters, and determine what they will be equal to in order to fit a set of data. An example is shown below.

```python
###### Import numpy and solver
import numpy as np
import scipy.optimize as opt

###### Define model

def model(x,*pars):
    
    # Indepdendent variables
    CA = x[0,:]
    CB = x[1,:]
    CC = x[2,:]
    ndtps = len(CA)
    
    # Parameters to be fit
    b1 = pars[0]
    b2 = pars[1]
    b3 = pars[2]
    b4 = pars[3]
    b5 = pars[4]
    
    # Calculate output value for each data point based on a model
    rate_model = (b1*CB-CC/b5)/(1+b2*CA+b3*CB+b4*CC)  # Manually write the rate expression
    return rate_model

rate_exp = [8.55,3.79,4.82,0.02,2.75,14.39,2.54,4.35,13,8.5,0.05,11.32,3.13]
CA_exp = [470,285,470,470,470,100,100,470,100,100,100,285,285]
CB_exp = [300,80,300,80,80,190,80,190,300,300,80,300,190]
CC_exp = [10,10,120,120,10,10,65,65,54,120,120,10,120]

###### Parameter guesses
guess = [1,0.1,0.1,0.1,1]

####### Regression
xvars = np.array([CA_exp,CB_exp,CC_exp])
popt,pcov = opt.curve_fit(model,xvars,rate_exp,guess,method='lm')

print('Optimized parameters: \n',popt,'\n')

###### Check solution
rate_predict = model(xvars,*popt)
print('Experimental values: \n',rate_exp)
print('Model values: \n',rate_predict)

###### Parity plot

x_parity = [min([min(rate_exp),min(rate_predict)]),max([max(rate_exp),max(rate_predict)])]
y_parity = [min([min(rate_exp),min(rate_predict)]),max([max(rate_exp),max(rate_predict)])]
    
fig = plt.figure()
ax1 = plt.subplot(111)

ax1.scatter(rate_exp,rate_predict,marker='o',edgecolors= 'black')
ax1.plot(x_parity,y_parity,'--')
ax1.set_xlabel('Experimental Rate')
ax1.set_ylabel('Predicted Rate')

fig.show()
```

#### solve_ivp: ODE solver

"solve_ivp" is the method of choice to integrate a system of ODEs. An example is below:

This sample has the following two systems of equations:

$$ x = k_1 \cdot x - k_2 \cdot x \cdot y $$
$$ y = k_3 \cdot x \cdot y - k_4 \cdot y $$

```python
from scipy.integrate import solve_ivp

def diff(t,U):
   
    # Assign values to inputs if desired
    x = U[0]
    y = U[1]
    
    # Define equations
    dU=np.zeros(len(U))
    dU[0] = k1 * x - k2 * x * y
    dU[1] = k3 * x * y - k4 * y
    
    # Reorganize list to fit required dimensions 
    dU = np.array(dU).tolist()
    
    return dU

# Define constants
k1 = 0.2     # days**-1
k2 = 0.00004  # days**-1 foxes**-1
k3 = 0.00004  # days**-1 rabbits**-1
k4 = 0.04     # days**-1

# Set initial conditions
U0 = [750,300]

# Specify integration limits
tlim = (0,800)
tlist = np.linspace(tlim[0],tlim[1],num=100)

# Integrate
solution = solve_ivp(diff,tlim,U0,t_eval=tlist)

# Obtain outputs and times
U = solution.y.T
t = solution.t
x = U[:,0]
y = U[:,1]

# Plot data
fig = plt.figure()
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)

# x and y vs t
ax1.plot(t,x,'b',label='x')
ax1.plot(t,y,'r',label='y')
ax1.set_xlabel('t')
ax1.set_ylabel('x or y')
ax1.set_title('x and y vs t')
ax1.legend()

# y vs x
ax2.plot(x,y,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('y vs x')

fig.tight_layout()
plt.show()
```
